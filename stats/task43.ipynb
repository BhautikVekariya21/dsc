# Probability Problems Solutions

## Binomial Distribution Problems (1-7)

### Problem 1: Email Marketing Click-Through Rate
**Given:** n = 100, p = 0.15, find P(X = 20)

Using the binomial probability formula: P(X = k) = C(n,k) × p^k × (1-p)^(n-k)

P(X = 20) = C(100,20) × (0.15)^20 × (0.85)^80

**Answer:** P(X = 20) ≈ 0.0419 or 4.19%

### Problem 2: Medication Success Rate
**Given:** n = 50, p = 0.75, find P(X < 35)

Since np = 37.5 > 5 and n(1-p) = 12.5 > 5, we can use normal approximation:
- μ = np = 37.5
- σ = √(np(1-p)) = √(9.375) ≈ 3.06

Using continuity correction: P(X < 35) = P(X ≤ 34.5)
Z = (34.5 - 37.5)/3.06 = -0.98

**Answer:** P(X < 35) ≈ 0.1635 or 16.35%

### Problem 3: Premium Subscription Sign-up Rate
**Given:** n = 500, p = 0.20, find P(90 ≤ X ≤ 110)

Normal approximation:
- μ = np = 100
- σ = √(np(1-p)) = √(80) ≈ 8.94

Using continuity correction: P(89.5 ≤ X ≤ 110.5)
Z₁ = (89.5 - 100)/8.94 = -1.17
Z₂ = (110.5 - 100)/8.94 = 1.17

**Answer:** P(90 ≤ X ≤ 110) ≈ 0.7580 or 75.80%

### Problem 4: Reading Program Success
**Given:** n = 200, p = 0.70, find P(X > 140)

Normal approximation:
- μ = np = 140
- σ = √(np(1-p)) = √(42) ≈ 6.48

Using continuity correction: P(X > 140) = P(X ≥ 140.5)
Z = (140.5 - 140)/6.48 = 0.077

**Answer:** P(X > 140) ≈ 0.4693 or 46.93%

### Problem 5: Electronic Component Defect Rate
**Given:** n = 200, p = 0.05, find P(X < 10)

Normal approximation:
- μ = np = 10
- σ = √(np(1-p)) = √(9.5) ≈ 3.08

Using continuity correction: P(X < 10) = P(X ≤ 9.5)
Z = (9.5 - 10)/3.08 = -0.16

**Answer:** P(X < 10) ≈ 0.4364 or 43.64%

### Problem 6: Ice Cream Preference (Bernoulli)
**Given:** p(chocolate) = 0.70, find p(vanilla)

For a single trial: P(vanilla) = 1 - P(chocolate) = 1 - 0.70 = 0.30

**Answer:** P(prefers vanilla) = 0.30 or 30%

### Problem 7: Software Bug Rate
**Given:** n = 10,000, p = 0.02, find P(X ≥ 250)

Normal approximation:
- μ = np = 200
- σ = √(np(1-p)) = √(196) = 14

Using continuity correction: P(X ≥ 250) = P(X ≥ 249.5)
Z = (249.5 - 200)/14 = 3.54

**Answer:** P(X ≥ 250) ≈ 0.0002 or 0.02%

## Normal Distribution and Central Limit Theorem (8-10)

### Problem 8: Adult Men Heights
**Given:** μ = 69.2 inches, σ = 2.9 inches, n = 50, find P(X̄ > 70)

Standard error: σ_x̄ = σ/√n = 2.9/√50 ≈ 0.41

Z = (70 - 69.2)/0.41 = 1.95

**Answer:** P(X̄ > 70) ≈ 0.0256 or 2.56%

### Problem 9: Employee Salaries (CLT)
**Given:** μ = $75,000, σ = $10,000, n = 100, find P(X̄ < $72,500)

Standard error: σ_x̄ = 10,000/√100 = 1,000

Z = (72,500 - 75,000)/1,000 = -2.5

**Answer:** P(X̄ < $72,500) ≈ 0.0062 or 0.62%

### Problem 10: Restaurant Wait Times (CLT)
**Given:** μ = 15 minutes, σ = 3 minutes, n = 50, find P(X̄ > 16)

Standard error: σ_x̄ = 3/√50 ≈ 0.424

Z = (16 - 15)/0.424 = 2.36

**Answer:** P(X̄ > 16) ≈ 0.0091 or 0.91%

## Key Formulas Used

### Binomial Distribution
- Mean: μ = np
- Standard deviation: σ = √(np(1-p))
- Normal approximation valid when np ≥ 5 and n(1-p) ≥ 5

### Central Limit Theorem
- Sample mean distribution: X̄ ~ N(μ, σ²/n)
- Standard error: σ_x̄ = σ/√n
- Z-score: Z = (X̄ - μ)/σ_x̄

### Continuity Correction
- For discrete distributions approximated by normal: adjust by ±0.5
- P(X ≤ k) → P(X ≤ k + 0.5)
- P(X ≥ k) → P(X ≥ k - 0.5)
- P(X < k) → P(X ≤ k - 0.5)
- P(X > k) → P(X ≥ k + 0.5)
