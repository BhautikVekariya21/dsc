### Colored Scatterplots

```python
iris = pd.read_csv('./iris.csv')
iris.sample(5)
```

```python
iris['species'] = iris['species'].replace({'setosa':0,'versicolor':1,'virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['sepal_length'],iris['petal_length'],c=iris['species'],cmap='summer',alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar(label='Species')
```

```python
plt.scatter(iris['sepal_length'],iris['petal_length'],c=iris['species'],cmap='viridis',alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar(label='Species')
```

## plot size

```python
plt.figure(figsize=(10, 6))
plt.scatter(iris['sepal_length'],iris['petal_length'],c=iris['species'],cmap='winter',alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Iris Dataset: Sepal Length vs Petal Length')
plt.colorbar(label='Species')
```

### 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx,yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python

fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p=ax.plot_surface(xx, yy, z, cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.log(xx)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

### Contour Plots

```python
fig = plt.figure(figsize=(8,8))
ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
