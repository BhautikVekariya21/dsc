### Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart
import matplotlib.pyplot as plt
childen = [10,22,21,32,14,41]
colors = ['red','blue','green','yellow','white','pink']
plt.bar(colors,childen,color=colors)
```

```python
# horizontal bar chart
plt.barh(colors,childen,color=colors)
```

```python
# color and label 
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2, color='#acc2f5', label='2015')
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2, color='#bd27e1', label='2016')
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2, color='#41d0bc', label='2017')
plt.legend()
plt.grid(True, linestyle='--', linewidth=0.5, color='black')
plt.title('Batsman Performance Over Seasons')
# Fix x-axis labels
plt.xticks(np.arange(df.shape[0]), df['batsman'], rotation=45)
```

```python
# a problem 
childern = [10,12,32,41,52,74,85,32,12]
color = ["red red red","green green green green","blueblue blue blue","yellow yellow yellow","orange orange orange","purple purple purple purple","pink pink pink","brown brown brown brown","black black black black black"]
plt.bar(color,childern,color='magenta')
plt.grid(True, linestyle='--', linewidth=0.5, color='black')
```

```python
# rotation
childern = [10,12,32,41,52,74,85,32,12]
color = ["red red red","green green green green","blueblue blue blue","yellow yellow yellow","orange orange orange","purple purple purple purple","pink pink pink","brown brown brown brown","black black black black black"]
plt.bar(color,childern,color='magenta')
plt.xticks(rotation=90)
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23,14,41,56,78,95]
subjects = ['eng','science','maths','socialscience','gujarati','sanskrit']
plt.pie(data,labels=subjects)
```

```python
# dataset
df = pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['#acc2f5','#0e8e91','#B6DF00','#27F302','#933AC6','#CBDF1BE0'])
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=["#528489","#b6b6b6",'#742ec3','#c355e6','#d8a138','#25cbe2'])
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```

```python
arr = np.load('big-array.npy')
plt.hist(arr,bins=[10,20,30,40,50,60,70],log=True)
plt.grid(True, linestyle='--', linewidth=0.5, color='black')
plt.title('Histogram of Big Array')
plt.xlabel('Value Ranges')
plt.ylabel('Frequency (Log Scale)')
plt.show()
```
