### Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart
import matplotlib.pyplot as plt
childen = [10,22,21,32,14,41]
colors = ['red','blue','green','yellow','white','pink']
plt.bar(colors,childen,color=colors)
```

```python
# horizontal bar chart
plt.barh(colors,childen,color=colors)
```

```python
# color and label 
df = pd.read_csv('./batsman_season_record.csv')
df
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2, color='yellow', label='2015')
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2, color='pink', label='2016')
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2, color='green', label='2017')

# Fix x-axis labels
plt.xticks(np.arange(df.shape[0]), df['batsman'], rotation=45)
```

```python
# a problem 
childern = [10,12,32,41,52,74,85,32,12]
color = ["red red red","green green green green","blueblue blue blue","yellow yellow yellow","orange orange orange","purple purple purple purple","pink pink pink","brown brown brown brown","black black black black black"]
plt.bar(color,childern,color='magenta')
plt.xticks(rotation=90)
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23,14,41,56,78,95]
subjects = ['eng','science','maths','socialscience','gujarati','sanskrit']
plt.pie(data,labels=subjects)
```

```python
# dataset
df = pd.read_csv('./gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','magenta','yellow','white','cyan','red'])
```

```python
df['batsman_runs'] = pd.to_numeric(df['batsman_runs'], errors='coerce').fillna(0)

# Ensure explode is the correct length
explode = [0.1] + [0] * (df.shape[0] - 1) if len(df) > 1 else [0.1]  # Avoids single-element error

# Plot pie chart
plt.pie(df['batsman_runs'], labels=df['batsman'], explode=explode, 
        autopct='%1.1f%%', shadow=True, startangle=90)
```
