```python
# %pip install plotly

import numpy as np
import plotly.express as px
import pandas as pd
import seaborn as sns
# import datasets
tips = px.data.tips()
iris = px.data.iris()
gap = px.data.gapminder()
```

```python
temp_df = gap[gap['year']==2007]
```

```python
import plotly.graph_objects as go

# Create the scatter plot
trace1 = go.Scatter(
    x=temp_df['lifeExp'],
    y=temp_df['gdpPercap'],
    mode='markers'
)

# Wrap trace in a list
data = [trace1]

# Define layout
layout = go.Layout(
    title='Life Exp vs GDP per capita for 2007',
    xaxis={'title': 'Life Exp'},
    yaxis={'title': 'GDP'}
)

# Corrected syntax
fig = go.Figure(data=data, layout=layout)

fig.show()
```

```python
px.scatter(
    temp_df, 
    x='lifeExp',  # X-axis: Life Expectancy for each country
    y='gdpPercap',  # Y-axis: GDP per Capita for each country
    color='continent',  # Color each point by the continent of the country
    size='pop',  # Adjust the size of each point based on the population of the country
    hover_name='country',  # Show the country's name when hovering over the point
    log_x=True,  # Apply a logarithmic scale to the X-axis (life expectancy)
    log_y=True,  # Apply a logarithmic scale to the Y-axis (GDP per capita)
    range_x=[temp_df['lifeExp'].min(), temp_df['lifeExp'].max()],  # Set the range for the X-axis based on the minimum and maximum values of life expectancy
    range_y=[temp_df['gdpPercap'].min(), temp_df['gdpPercap'].max()],  # Set the range for the Y-axis based on the minimum and maximum values of GDP per capita
    template="plotly_dark"  # Use a dark theme for the plot
)
```

```python
# plot animation of the above curve on the basic if the year
import plotly.express as px

px.scatter(gap, 
           x='lifeExp', 
           y='gdpPercap',
           animation_frame='year',  # Animate the plot frame-by-frame by year
           animation_group='country',  # Group data by country for animation
           size='pop',  # Size of the points based on population
           color='continent',  # Color points based on continent
           hover_name='country',  # Show the country name when hovering
           size_max=50,  # Max size of the points
           template='plotly_dark',
           range_x=[0, 100]  # Set the x-axis range to [0, 100]
)
# You're telling Plotly which data points across animation frames belong to the same entity (like the same country over time). 
# This allows Plotly to track each point smoothly across frames rather than treating each frame as a completely separate scatter plot.
```

```python
# plot pie chart of world pop of europian coutries in 2007 -> explode(pull)
# no support for pull in plotly express 
import plotly.graph_objects as go

labels = ['Oxygen','Hydrogen','Carbon_Dioxide','Nitrogen']
values = [4500, 2500, 1053, 500]

# pull is given as a fraction of the pie radius
fig = go.Figure(data=[go.Pie(labels=labels, values=values, pull=[0, 0, 0.2, 0])])
fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Define the grid
x = np.linspace(19, 503, 100)
y = np.linspace(19, 503, 100)
xx, yy = np.meshgrid(x, y)

# Funky mathematical function
z = np.sin(xx) + np.log(xx)

# Create surface plot
trace = go.Surface(x=x, y=y, z=z)

# Define layout with dark theme
layout = go.Layout(
    template='plotly_dark',  # Dark background
)

# Create and show figure
fig = go.Figure(data=[trace], layout=layout)
fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Define the grid
x = np.linspace(21, 741, 412)
y = np.linspace(21, 741, 412)
xx, yy = np.meshgrid(x, y)

# Funky mathematical function
z =  xx*np.log(yy)

# Create surface plot
trace = go.Surface(x=x, y=y, z=z)

# Define layout with dark theme
layout = go.Layout(
    title='Weird 3D Surface Plot',
    template='plotly_dark',  # Dark background
)

# Create and show figure
fig = go.Figure(data=[trace], layout=layout)
fig.show()
```

```python
# 3d surface plot 
# can not plot 3 d surface plot using plotly express 
# we will use for 3 d surfase plotly go
import numpy as np
import plotly.graph_objects as go

# Define the grid
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
xx, yy = np.meshgrid(x, y)

# Funky mathematical function
z = np.sin(xx) * np.cos(yy) * np.exp(-0.1 * (xx**2 + yy**2))

# Create surface plot
trace = go.Surface(x=x, y=y, z=z)

# Define layout with dark theme
layout = go.Layout(
    title='Weird 3D Surface Plot',
    template='plotly_dark',  # Dark background
)

# Create and show figure
fig = go.Figure(data=[trace], layout=layout)
fig.show()
```

```python
#contour plot
# can not plot contour plot using plotly express 
# we will use for contour plotly go
import numpy as np
import plotly.graph_objects as go

# Define the grid
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
xx, yy = np.meshgrid(x, y)

# Funky mathematical function
z = np.sin(xx) * np.cos(yy) * np.exp(-0.1 * (xx**2 + yy**2))

# Create surface plot
trace = go.Contour(x=x, y=y, z=z)

# Define layout with dark theme
layout = go.Layout(
    title='Weird 3D Surface Plot',
    template='plotly_dark',  # Dark background
)

# Create and show figure
fig = go.Figure(data=[trace], layout=layout)
fig.show()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.express as px

# Create subplots (1 row, 2 columns)
fig = make_subplots(rows=1, cols=2)

# Add scatter plot (Extracting the first trace from px.scatter)
scatter_fig = px.scatter(x=[1, 9, 8, 17], y=[2, 10, 11, 13])
fig.add_trace(scatter_fig.data[0], row=1, col=1)

# Add histogram
fig.add_trace(
    go.Histogram(x=[1, 2, 8, 44, 15, 74, 136, 78, 95, 95, 74, 12, 32, 42, 6, 32, 41, 25, 71]),
    row=1,
    col=2
)

# Update layout with dark theme
fig.update_layout(title_text="Side By Side Subplots", template="plotly_dark")

# Show figure
fig.show()
```
