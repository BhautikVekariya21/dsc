```python
temp_df = gap[gap['year']==2007]
```

```python
import plotly.graph_objects as go

# Create the scatter plot
trace1 = go.Scatter(
    x=temp_df['lifeExp'],
    y=temp_df['gdpPercap'],
    mode='markers'
)

# Wrap trace in a list
data = [trace1]

# Define layout
layout = go.Layout(
    title='Life Exp vs GDP per capita for 2007',
    xaxis={'title': 'Life Exp'},
    yaxis={'title': 'GDP'}
)

# Corrected syntax
fig = go.Figure(data=data, layout=layout)

fig.show()
```

```python
# plot life exp and gdp scatter plot -> continent as color -> pop as size -> hover name -> range_x/range_y -> log_x/log_y
px.scatter(
    temp_df, 
    x='lifeExp', 
    y='gdpPercap',  # Corrected column name
    color='continent',  # Assigns colors based on continent
    size='pop',  # Point size based on population
    hover_name='country',  # Shows country name on hover
    log_x=True,  # Log scale for better visualization
    log_y=True,  # Log scale for GDP per capita
    range_x=[temp_df['lifeExp'].min(), temp_df['lifeExp'].max()],  # Set X-axis range
    range_y=[temp_df['gdpPercap'].min(), temp_df['gdpPercap'].max()],  # Set Y-axis range
    template="plotly_dark"
)

```

```python
# plot animation of the above curve on the basic if the year

px.scatter(gap,x='lifeExp',y='gdpPercap',
           animation_frame='year',animation_group='country',
           size='pop',color='continent',hover_name='country',
           size_max=50,template='plotly_dark')
```

```python
# line plot
# plot india pop line plot

india = gap[gap['country'] == 'India']
px.line(india,x='year',y='pop',title='india pop',template='plotly_dark')
```

```python
# plot pie chart of world pop of europian coutries in 2007 -> explode(pull)
# no support for pull in plotly express 
import plotly.graph_objects as go

labels = ['Oxygen','Hydrogen','Carbon_Dioxide','Nitrogen']
values = [4500, 2500, 1053, 500]

# pull is given as a fraction of the pie radius
fig = go.Figure(data=[go.Pie(labels=labels, values=values, pull=[0, 0, 0.2, 0])])
fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Define the grid
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
xx, yy = np.meshgrid(x, y)

# Funky mathematical function
z = np.sin(xx) + np.log(xx)

# Create surface plot
trace = go.Surface(x=x, y=y, z=z)

# Define layout with dark theme
layout = go.Layout(
    template='plotly_dark',  # Dark background
)

# Create and show figure
fig = go.Figure(data=[trace], layout=layout)
fig.show()

```

```python
# 3d surface plot 
# can not plot 3 d surface plot using plotly express 
# we will use for 3 d surfase plotly go
import numpy as np
import plotly.graph_objects as go

# Define the grid
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
xx, yy = np.meshgrid(x, y)

# Funky mathematical function
z = np.sin(xx) * np.cos(yy) * np.exp(-0.1 * (xx**2 + yy**2))

# Create surface plot
trace = go.Surface(x=x, y=y, z=z)

# Define layout with dark theme
layout = go.Layout(
    title='Weird 3D Surface Plot',
    template='plotly_dark',  # Dark background
)

# Create and show figure
fig = go.Figure(data=[trace], layout=layout)
fig.show()
```

```python
#contour plot
# can not plot contour plot using plotly express 
# we will use for contour plotly go
import numpy as np
import plotly.graph_objects as go

# Define the grid
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
xx, yy = np.meshgrid(x, y)

# Funky mathematical function
z = np.sin(xx) * np.cos(yy) * np.exp(-0.1 * (xx**2 + yy**2))

# Create surface plot
trace = go.Contour(x=x, y=y, z=z)

# Define layout with dark theme
layout = go.Layout(
    title='Weird 3D Surface Plot',
    template='plotly_dark',  # Dark background
)

# Create and show figure
fig = go.Figure(data=[trace], layout=layout)
fig.show()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.express as px

# Create subplots (1 row, 2 columns)
fig = make_subplots(rows=1, cols=2)

# Add scatter plot (Extracting the first trace from px.scatter)
scatter_fig = px.scatter(x=[1, 9, 8, 17], y=[2, 10, 11, 13])
fig.add_trace(scatter_fig.data[0], row=1, col=1)

# Add histogram
fig.add_trace(
    go.Histogram(x=[1, 2, 8, 44, 15, 74, 136, 78, 95, 95, 74, 12, 32, 42, 6, 32, 41, 25, 71]),
    row=1,
    col=2
)

# Update layout with dark theme
fig.update_layout(title_text="Side By Side Subplots", template="plotly_dark")

# Show figure
fig.show()
```

```python
import plotly.express as px
import pandas as pd

# Data for the Gantt chart (LMS project timeline)
data = [
    dict(Task="Project Initialization & Setup", Start="2025-01-25", Finish="2025-02-01", Resource="Planning"),
    dict(Task="Frontend Development (React)", Start="2025-02-02", Finish="2025-02-25", Resource="Frontend"),
    dict(Task="Backend Development (Node.js, Express, Mongoose)", Start="2025-02-08", Finish="2025-03-05", Resource="Backend"),
    dict(Task="Database Setup (MongoDB)", Start="2025-02-15", Finish="2025-02-28", Resource="Database"),
    dict(Task="Security Integration (Bcrypt)", Start="2025-02-25", Finish="2025-03-10", Resource="Security"),
    dict(Task="Version Control Setup (Git)", Start="2025-03-05", Finish="2025-03-15", Resource="Version Control"),
    dict(Task="API Testing (Postman)", Start="2025-03-10", Finish="2025-03-25", Resource="Testing"),
    dict(Task="Unit Testing & Debugging", Start="2025-03-20", Finish="2025-04-10", Resource="Testing"),
    dict(Task="Final Deployment & Documentation", Start="2025-04-05", Finish="2025-04-20", Resource="Deployment"),
    dict(Task="Project Review & Closure", Start="2025-04-15", Finish="2025-04-25", Resource="Closure"),
]

# Create a DataFrame
df = pd.DataFrame(data)

# Create the Gantt chart using Plotly Express
fig = px.timeline(
    df,
    x_start="Start",
    x_end="Finish",
    y="Task",
    color="Resource",  # Color-code tasks by resource
    title="Gantt Chart for LMS Project (Jan 25  Apr 25, 2025)",
)

# Update layout for better readability
fig.update_yaxes(categoryorder="total ascending")  # Sort tasks in ascending order
fig.update_layout(
    xaxis_title="Timeline",
    yaxis_title="Tasks",
    showlegend=True,
    height=600,
    width=1000,
    title_x=0.5,  # Center the title
)

# Show the Gantt chart
fig.show()
```
