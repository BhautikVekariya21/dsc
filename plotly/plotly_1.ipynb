```python
# import the libraries
%pip install --upgrade nbformat


import plotly.graph_objects as go
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
# import datasets
tips = px.data.tips()
iris = px.data.iris()
gap = px.data.gapminder()
```

```python
temp_df = gap[gap['year']==2007]
```

```python
import plotly.graph_objects as go

# Create the scatter plot
trace1 = go.Scatter(
    x=temp_df['lifeExp'],
    y=temp_df['gdpPercap'],
    mode='markers'
)

# Wrap trace in a list
data = [trace1]

# Define layout
layout = go.Layout(
    title='Life Exp vs GDP per capita for 2007',
    xaxis={'title': 'Life Exp'},
    yaxis={'title': 'GDP'}
)

# Corrected syntax
fig = go.Figure(data=data, layout=layout)

fig.show()
```

```python
# plot life exp and gdp scatter plot -> continent as color -> pop as size -> hover name -> range_x/range_y -> log_x/log_y
px.scatter(
    temp_df, 
    x='lifeExp', 
    y='gdpPercap',  # Corrected column name
    color='continent',  # Assigns colors based on continent
    size='pop',  # Point size based on population
    hover_name='country',  # Shows country name on hover
    log_x=True,  # Log scale for better visualization
    log_y=True,  # Log scale for GDP per capita
    range_x=[temp_df['lifeExp'].min(), temp_df['lifeExp'].max()],  # Set X-axis range
    range_y=[temp_df['gdpPercap'].min(), temp_df['gdpPercap'].max()],  # Set Y-axis range
    template="plotly_dark"
)

```

```python
# plot animation of the above curve on the basic if the year

px.scatter(gap,x='lifeExp',y='gdpPercap',
           animation_frame='year',animation_group='country',
           size='pop',color='continent',hover_name='country',
           size_max=50,template='plotly_dark')
```

```python
# line plot
# plot india pop line plot

india = gap[gap['country'] == 'India']
px.line(india,x='year',y='pop',title='india pop',template='plotly_dark')
```

```python
# plot india,china and pak line plot
indchpk = gap[gap['country'].isin(['India','China','Pakistan'])].pivot(index='year', columns='country', values='lifeExp')
```

```python
px.line(indchpk,x=indchpk.index,y=indchpk.columns,template='plotly_dark')
```

```python
#bar chart 
# india's pop over the years
px.bar(india,x='year',y='pop',template='plotly_dark',text='pop',text_auto=True)
```

```python
#pop comp of 3 countries
indcp = gap[gap['country'].isin(['India', 'China', 'Pakistan'])].pivot(index='year', columns='country', values='gdpPercap')
px.bar(indcp, x=indcp.index, y=indcp.columns,barmode='group',template='plotly_dark',text_auto=True)
```

```python
#grouped bar chsrt -> text_auto
indcp = gap[gap['country'].isin(['India', 'China', 'Pakistan'])].pivot(index='year', columns='country', values='pop')
px.bar(indcp, x=indcp.index, y=indcp.columns,barmode='group',template='plotly_dark',log_y=True,text_auto=True)
```

```python
#stacked bar chart
# pop contribution per country to a cotinents pop stacked for a particular year(2007)
temp_df = gap[gap['year']==2007]
px.bar(temp_df,x='continent',y='pop',color='country',template='plotly_dark',text_auto=True)
```

```python
# bar chart animation
px.bar(
    data_frame=gap,  # <-- Add your DataFrame here
    x='continent',  # X-axis as continent
    y='pop',  # Y-axis as population
    color='continent',  # Color bars based on continent
    animation_frame='year',  # Create animation over years
    animation_group='country',  # Group animation by country
    template='plotly_dark',  # Dark theme
    range_y=[0, 4000000000]  # Range of y-axis
)
```

```python
# histogram
# plot histogram of life expt of all continents pop stacked for a particular year(2007)
temp_df = gap[gap['year']==2007]
px.histogram(temp_df,x='lifeExp',template='plotly_dark',nbins=10,text_auto=True)
```

```python
# plot histogram of life expt of all countries in 2007 -> nbins text_auto
df_2007 = gap[gap['year'] == 2007]

# Create histogram
px.histogram(
    data_frame=df_2007,  # Filtered DataFrame
    x="lifeExp",  # Life Expectancy on X-axis
    nbins=30,  # Number of bins
    text_auto=True,  # Show counts on bars
    template="plotly_dark"  # Dark theme
)
```

```python
# plot histogram of sepal length of all iris species
px.histogram(iris,x='sepal_length',color='species',nbins=30,text_auto=True,template='plotly_dark')
```

```python
# pie -> values -> names 
# find the pie chart of europian countries in 2007
temp_df = gap[(gap['year'] == 2007) & (gap['continent']=='Europe')]

px.pie(temp_df,values='pop',names='country',title='Population of European countries in 2007',template='plotly_dark')
```

```python
px.pie(tips, values='tip', names='day', color='day',template='plotly_dark',
             color_discrete_map={'Thur':'lightcyan',
                                 'Fri':'cyan',
                                 'Sat':'royalblue',
                                 'Sun':'darkblue'})
```

```python
# plot pie chart of world pop of europian coutries in 2007 -> explode(pull)
# no support for pull in plotly express 
import plotly.graph_objects as go

labels = ['Oxygen','Hydrogen','Carbon_Dioxide','Nitrogen']
values = [4500, 2500, 1053, 500]

# pull is given as a fraction of the pie radius
fig = go.Figure(data=[go.Pie(labels=labels, values=values, pull=[0, 0, 0.2, 0])])
fig.show()
```

```python
# sunburst plot -> sunburnt plots visulize hirarechal data spanning outwards radially from root to leaves. -> color
# path -> [],values
temp_df = gap[gap['year'] == 2007]
px.sunburst(temp_df, path=['continent', 'country'], values='pop',template='plotly_dark',color='lifeExp')
```

```python
px.sunburst(tips,path=['sex','smoker','day','time'],values='total_bill',color='size',template='plotly_dark')
```

```python
# treemap
temp_df = gap[gap['year'] == 2007]
px.treemap(temp_df, path = [px.Constant('world'), 'continent', 'country'], values = 'pop',
           color = 'lifeExp', hover_data = ['pop'], color_continuous_scale = 'RdBu',template='plotly_dark')
```

```python
#heatmap -> find heatmap of all continents with year on avg life exp
temp_df = gap.pivot_table(index='year',columns='continent',values='lifeExp',aggfunc='mean')
px.imshow(temp_df,template='plotly_dark')
```

```python
#3d scatterplot
# plot a 3d scatter plot of all country data for 2007
temp_df = gap[gap['year'] == 2007]
px.scatter_3d(temp_df,x='lifeExp',y='pop',z='gdpPercap',color='continent',size='pop',size_max=60,hover_name='country',log_x=True,log_y=True,log_z=True,template='plotly_dark')
```

```python
px.scatter_3d(iris,x='sepal_length',y='sepal_width',z='petal_length',color='species',opacity=0.7,template='plotly_dark')
```

```python
#scatter_matrix -> dimensions
px.scatter_matrix(iris, dimensions = ["sepal_width", "sepal_length", "petal_width", "petal_length"], color = "species",template='plotly_dark')
```
