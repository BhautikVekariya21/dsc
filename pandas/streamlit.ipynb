```python
import pandas as pd
df = pd.read_csv('indian_startup_funding_2023.csv')
```

```python
df.drop(columns=['Remarks'],inplace=True)
```

```python
df.head()
```

```python
df.set_index('S.No',inplace=True)
```

```python
df.rename(columns={
    'Date dd/mm/yyyy':'date',
    'Startup Name':'startup',
    'Industry Vertical':'vertical',
    'SubVertical':'subvertical',
    'City  Location':'city',
    'Investors Name':'investors',
    'Amount in USD':'amount',
    'Investment Type':'round',

},inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['amount'] = df['amount'].fillna('0')
```

```python
df['amount'] = df['amount'].apply(lambda x: x.replace('$', '')
                                      .replace(',', '')
                                      .replace('undisclosed', '0')
                                      .replace('unknown', '0')
                                      .replace('Undisclosed', '0')
                                      .replace('14342000+', '14342000')
                                      .replace('\\\\xc2\\\\xa020000000', '0')
                                      .replace('\\\\xc2\\\\xa016200000', '0')
                                      .replace('\\\\xc2\\\\xa0N/A', '0')
                                      .replace('\\\\xc2\\\\xa0', '0')
                              ).astype(float)
```

```python
df.head()
```

```python
df.info()
```

```python
df['amount'] = df['amount'] / 1e6
```

```python
df.head()
```

```python
df['date'] = df['date'].replace('05/072018', '07/05/2018')
```

```python

df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y', errors='coerce')
```

```python
df.head()
```

```python
df = df.dropna(subset=['date','startup','vertical','city','investors','amount'])
```

```python
df.info()
```

```python
df.to_csv('startup_cleaned.csv',index=False)
```
