```python
# agg method
# passing dict
genres.agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

```python
# Collect DataFrames in a list
dfs = []
for group, data in genres:
    dfs.append(data[data['IMDB_Rating'] == data['IMDB_Rating'].max()])

# Concatenate all DataFrames
df = pd.concat(dfs, ignore_index=True)
df
```

```python
# split (apply) combine
# apply -> builtin function

genres.apply(np.minimum.reduce, include_groups=False)
```

```python
# find number of movies starting with A for each group

def foo(group):
  return group['Series_Title'].str.startswith('A').sum()
genres.apply(foo, include_groups=False)
```

```python
def rank_movie(group):
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group
genres.apply(rank_movie,include_groups=False)
```

```python
# find normalized IMDB rating group wise

def normal(group):
  group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
  return group

genres.apply(normal,include_groups=False)
```

```python
import pandas as pd
import numpy as np
ipl = pd.read_csv('deliveries.csv')
ipl
```

```python
ipl.info()
```

```python
# find the top 10 batsman in terms of runs
ipl.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max no of sixes
six = ipl[ipl['batsman_runs'] == 6]

six.groupby('batter')['batter'].count().sort_values(ascending=False).head(1).index[0]
```

During internal calculations, like groupby().count() or groupby().sum(), pandas uses the grouping column as a regular column internally (so calculations like ['batsman'].count() work).

But in the output, the grouping column becomes the index, not a regular column. So, if you want it as a regular column in the result, you need to use .reset_index().

```python
# find batsman with most number of 4's and 6's in last 5 overs
temp_df = ipl[ipl['over'] > 15]
temp_df = temp_df[(temp_df['batsman_runs'] == 4) | (temp_df['batsman_runs'] == 6)]
temp_df.groupby('batter')['batter'].count().sort_values(ascending=False).head(1).index[0]
```

```python
# Step 1: Filter only last 5 overs
temp_df = ipl[ipl['over'] > 15]

# Step 2: Keep only 4s and 6s
temp_df = temp_df[(temp_df['batsman_runs'] == 4) | (temp_df['batsman_runs'] == 6)]

# Step 3: Group by batsman and sum the batsman_runs
runs_by_batsman = temp_df.groupby('batter')['batsman_runs'].sum()

# Step 4: Find the batsman with the highest total runs from 4s and 6s
top_batsman = runs_by_batsman.sort_values(ascending=False).head(1).index[0]

print(top_batsman)
```

```python
# find V Kohli's record against all teams
temp_df = ipl[ipl['batter'] == 'V Kohli']

temp_df.groupby('bowling_team')['batsman_runs'].sum().reset_index().sort_values('batsman_runs', ascending=False)
```

```python
# find V Kohli's record against all teams
temp_df = ipl[ipl['batter'] == 'V Kohli']

temp_df.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).reset_index()
```

```python
# Create a function that can return the highest score of any batsman
temp_df = ipl[ipl['batter'] == 'V Kohli']
temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
def highest(batsman):
  temp_df = ipl[ipl['batter'] == batsman]
  return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```
