### `Q-4:` Perform `groupby` based on the `"Team"` column and then perform Z Normalization on top of the below columns of each group:
1. Passes
2. Passes Completed
3. Attempted Line Breaks
4. Completed Line Breaks

You have to make a python function named `z_normalization` which takes two arguments:

1. *group:* Every group that you have created
2. *cols_to_perform:* This parameter takes a list of columns on which you have to perform the Z-Normalization.

For this problem, you have to use th `apply()` method.

$\Large Z \ - \ Normalization = \frac{X_i - \mu}{std} $

After that find out the below values for each group:
- minimum "Passess"
- maximum "Passess"
- minimum "Yellow Cards"
- maximum "Yellow cards"
- average "Yellow Cards"
- maximum "Attempted Line Breaks"
- minimum "Attempted Line Breaks"
- standardard deviation of "Attempted Line Breaks"
- average Possession

````python
import pandas as pd

# Load dataset
fifa_df = pd.read_csv("Fifa Worldcup 2022 - Sheet1.csv")

# Z-normalization function
def z_normalization(group, cols):
    for col in cols:
        mean = group[col].mean()
        std = group[col].std()
        group[col] = (group[col] - mean) / std if std != 0 else 0
    return group

# Columns to normalize
cols = ['Passes', 'Passes Completed', 'Attempted Line Breaks', 'Completed Line Breaks']

# Apply Z-normalization
norm_df = fifa_df.groupby('Team')[cols].apply(lambda x: z_normalization(x.copy(), cols), include_groups=False)

# Compute statistics
stats = fifa_df.groupby('Team').agg({
    'Passes': ['min', 'max'],
    'Yellow Cards': ['min', 'max', 'mean'],
    'Attempted Line Breaks': ['max', 'min', 'std'],
    'Possession (%)': 'mean'
}).round(2)

# Rename columns
stats.columns = ['Min Passes', 'Max Passes', 'Min Yellow Cards', 'Max Yellow Cards', 'Avg Yellow Cards',
                 'Max Line Breaks', 'Min Line Breaks', 'Std Line Breaks', 'Avg Possession']

# Print results
norm_df
````

### `Q-5:` Find batsman in below category-
* Highest score while chasing
* Best Strike rate while chasing and have faced 100+ balls


> Chasing mean team batting in second inning

````python
import pandas as pd
# Load dataset
df = pd.read_csv("deliveries.csv")
# Filter for chasing (second innings)
chasing_df = df[df['inning'] == 2]

# 1. Highest score while chasing
scores = chasing_df.groupby(['match_id', 'batter'])['batsman_runs'].sum().reset_index()
highest_score = scores.loc[scores['batsman_runs'].idxmax()]

# 2. Best strike rate while chasing (100+ balls faced)
balls_faced = chasing_df.groupby('batter')['ball'].count()
runs_scored = chasing_df.groupby('batter')['batsman_runs'].sum()
valid_batsmen = balls_faced[balls_faced >= 100].index
strike_rates = (runs_scored[valid_batsmen] / balls_faced[valid_batsmen] * 100).round(2)
best_strike_rate = strike_rates.idxmax()
best_strike_value = strike_rates.max()

print("Highest Score While Chasing:")
print(f"Batsman: {highest_score['batter']}, Score: {highest_score['batsman_runs']}, Match ID: {highest_score['match_id']}")
print("\nBest Strike Rate While Chasing (100+ balls faced):")
print(f"Batsman: {best_strike_rate}, Strike Rate: {best_strike_value}")
````

### `Q-6` Most Successful bowler against any batsman. Find that pair of bowler and batsman.
> Most Successful in terms of dissmissal. A bowler who have dissmissed any batsman most no of times. If any two pairs have same no of dissmisal, consider runs conceded by bowler to that batsman. Those who have concede lesser runs is more successful.

````python
df
````

````python
import pandas as pd

# Load dataset
df = pd.read_csv("deliveries.csv")

# 1. Calculate dismissals (excluding run outs)
dismissals = df[df['player_dismissed'].notna() & (df['dismissal_kind'] != 'run out')]
dismissal_counts = dismissals.groupby(['bowler', 'player_dismissed']).agg(
    dismissals=('player_dismissed', 'count')
).reset_index()

# 2. Calculate runs conceded from full data
run_stats = df.groupby(['bowler', 'batter']).agg(
    runs=('batsman_runs', 'sum')
).reset_index()

# 3. Merge on bowler and batsman/player_dismissed
stats = pd.merge(
    dismissal_counts,
    run_stats,
    left_on=['bowler', 'player_dismissed'],
    right_on=['bowler', 'batter'],
    how='left'
).drop(columns='batter')

# 4. Sort by most dismissals, and if tie, fewest runs
top_pair = stats.sort_values(['dismissals', 'runs'], ascending=[False, True]).iloc[0]

# 5. Print result
print(f"Bowler: {top_pair['bowler']}, Batsman: {top_pair['player_dismissed']}, "
      f"Dismissals: {top_pair['dismissals']}, Runs Conceded: {top_pair['runs']}")
````

### `Q-7`: Most successful batting pair in IPL. Batting pair who have scored most runs playing together.


````python
import pandas as pd

# Load dataset
df = pd.read_csv("deliveries.csv")

# Create a new column with sorted batsman & non-striker pair
df['pair'] = df[['batter', 'non_striker']].apply(lambda x: ' & '.join(sorted(x)), axis=1)

# Group by the pair and sum the runs
pair_runs = df.groupby('pair')['batsman_runs'].sum()

# Get the top pair (the one with the most runs)
top_pair = pair_runs.idxmax()
runs = pair_runs.max()

# Print the result
print(f"Batsman Pair: {top_pair}, Runs: {runs}")
````

### `Q-8:` Make a dataframe for all batting pairs played together.
```
Batsman1 Batsman2 Runs Avg StrikeRate
```

> Just to ease this question you can count wide-balls for strike rate.

````python
import pandas as pd

# Load dataset
df = pd.read_csv("deliveries.csv")

# Create sorted batsman pair
df['pair'] = df[['batsman', 'non_striker']].apply(lambda x: ' & '.join(sorted(x)), axis=1)

# Group by pair to aggregate stats
stats = df.groupby('pair').agg(
    Runs=('batsman_runs', 'sum'),
    Balls=('ball', 'count'),  # Legal deliveries
    Wides=('wide_runs', 'sum'),  # Count wide deliveries
    Dismissals=('player_dismissed', 'count')
).reset_index()

# Add wide balls to total balls
stats['TotalBalls'] = stats['Balls'] + stats['Wides']

# Compute Strike Rate and Average
stats['StrikeRate'] = (stats['Runs'] / stats['TotalBalls'] * 100).round(2)
stats['Avg'] = (stats['Runs'] / stats['Dismissals'].clip(lower=1)).round(2)  # Avoid division by zero

# Split pair into Batsman1 and Batsman2
stats[['Batsman1', 'Batsman2']] = stats['pair'].str.split(' & ', expand=True)

# Select final columns
batting_pairs = stats[['Batsman1', 'Batsman2', 'Runs', 'Avg', 'StrikeRate']]
# Create three sorted DataFrames
pairs_by_runs = batting_pairs.sort_values('Runs', ascending=False)
pairs_by_strike = batting_pairs.sort_values('StrikeRate', ascending=False)
pairs_by_avg = batting_pairs.sort_values('Avg', ascending=False)
# Combine the three sorted DataFrames into a single DataFrame

# Print result
batting_pairs
````
