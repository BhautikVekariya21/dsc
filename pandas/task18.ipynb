````python
import pandas as pd

# Load IPL dataset
df = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vRy2DUdUbaKx_Co9F0FSnIlyS-8kp4aKv_I0-qzNeghiZHAI_hw94gKG22XTxNJHMFnFVKsO4xWOdIs/pub?gid=1655759976&single=true&output=csv")

# Standardize team names
team_names = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Rising Pune Supergiants': 'Rising Pune Supergiant'
}
df[['Team1', 'Team2', 'WinningTeam', 'TossWinner']] = df[['Team1', 'Team2', 'WinningTeam', 'TossWinner']].replace(team_names)

# Exclude no-result matches
df = df[df['WinningTeam'].notna()]

# Get unique teams
teams = pd.concat([df['Team1'], df['Team2']]).unique()

# Calculate stats
stats = []
for team in teams:
    matches = df[df['Team1'] == team].shape[0] + df[df['Team2'] == team].shape[0]
    wins = df[df['WinningTeam'] == team].shape[0]
    home_matches = df[df['Team1'] == team].shape[0]
    home_wins = df[(df['Team1'] == team) & (df['WinningTeam'] == team)].shape[0]
    away_matches = df[df['Team2'] == team].shape[0]
    away_wins = df[(df['Team2'] == team) & (df['WinningTeam'] == team)].shape[0]
    
    stats.append({
        'Team': team,
        'Matches': matches,
        'Win%': round(100 * wins / matches, 2) if matches > 0 else 0,
        'Home Win%': round(100 * home_wins / home_matches, 2) if home_matches > 0 else 0,
        'Away Win%': round(100 * away_wins / away_matches, 2) if away_matches > 0 else 0
    })

# Create and sort DataFrame
result = pd.DataFrame(stats).sort_values('Win%', ascending=False)

# Print result
result
````

### `Q-9:` IPL Point Table

Make a function `point_table` which take `season` as parameter and show points table in non-ascendng order of points and in ascending order of team name.

For winning - 2 Ponits;
For loosing - 0 Point
For not result both team gets 1 points.

Make dataframe which will have
`TeamName` `MatchesPlayed` `MatchesWon` `NoResult` `Points`
make `TeamName` as index.

```
season parametr should be one of these->
['2022', '2021', '2020/21', '2019', '2018', '2017', '2016', '2015', '2014', '2013', '2012', '2011', '2009/10', '2009', '2007/08']
```


````python
import pandas as pd

def point_table(season):
    # Load IPL dataset
    df = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vRy2DUdUbaKx_Co9F0FSnIlyS-8kp4aKv_I0-qzNeghiZHAI_hw94gKG22XTxNJHMFnFVKsO4xWOdIs/pub?gid=1655759976&single=true&output=csv")

    # Standardize team names
    df[['Team1', 'Team2', 'WinningTeam']] = df[['Team1', 'Team2', 'WinningTeam']].replace({
        'Delhi Daredevils': 'Delhi Capitals',
        'Kings XI Punjab': 'Punjab Kings',
        'Rising Pune Supergiants': 'Rising Pune Supergiant'
    })

    # Filter season matches
    df = df[df['Season'] == season]
    if df.empty:
        return f"No matches for season {season}"

    # Get teams
    teams = pd.concat([df['Team1'], df['Team2']]).unique()

    # Calculate stats
    stats = []
    for team in teams:
        matches = df[df['Team1'] == team].shape[0] + df[df['Team2'] == team].shape[0]
        wins = df[df['WinningTeam'] == team].shape[0]
        no_result = df[((df['Team1'] == team) | (df['Team2'] == team)) & df['WinningTeam'].isna()].shape[0]
        stats.append({'Team': team, 'Matches': matches, 'Wins': wins, 'NoResult': no_result, 'Points': wins * 2 + no_result})

    # Create and sort DataFrame
    return pd.DataFrame(stats).set_index('Team').sort_values(['Points', 'Team'], ascending=[False, True])

# Example usage
print(point_table('2016'))
````

````python
import pandas as pd

def point_table(season, top_n=2):
    # Load IPL dataset
    df = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vRy2DUdUbaKx_Co9F0FSnIlyS-8kp4aKv_I0-qzNeghiZHAI_hw94gKG22XTxNJHMFnFVKsO4xWOdIs/pub?gid=1655759976&single=true&output=csv")

    # Filter season matches
    df = df[df['Season'] == season]
    if df.empty:
        return f"No matches for {season}"

    # Get teams
    teams = pd.concat([df['Team1'], df['Team2']]).unique()

    # Calculate stats
    stats = []
    for team in teams:
        matches = df[df['Team1'] == team].shape[0] + df[df['Team2'] == team].shape[0]
        wins = df[df['WinningTeam'] == team].shape[0]
        stats.append({'Team': team, 'Matches': matches, 'Wins': wins, 'Points': wins * 2})

    # Create and sort DataFrame, return top n
    return pd.DataFrame(stats).set_index('Team').sort_values(['Points', 'Team'], ascending=[False, True]).head(top_n)

# Print top 2 teams for 2022
print(point_table('2022', top_n=2))
````

### `Q-10:` IPL Point Table cont.
Extend the above IPL Point Table with an extra column as `SeasonPosition`

Team below top 4 after sorting on `Points` and then on `TeamName` Will have same `SeasonPosition` as there rank. use rank function.

Teams in Top four will have `SeasonPosition` as:

    'Winner' - Team won final

Note: If you try to chnage value of view of any dataframe a warnig will be shown. To avoid it, make a copy of the dataframe you want to change in by `df.copy()`

````python
import pandas as pd

def point_table(season):
    # Load IPL dataset
    df = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vRy2DUdUbaKx_Co9F0FSnIlyS-8kp4aKv_I0-qzNeghiZHAI_hw94gKG22XTxNJHMFnFVKsO4xWOdIs/pub?gid=1655759976&single=true&output=csv")

    # Standardize team names
    df[['Team1', 'Team2', 'WinningTeam']] = df[['Team1', 'Team2', 'WinningTeam']].replace({
        'Delhi Daredevils': 'Delhi Capitals',
        'Kings XI Punjab': 'Punjab Kings',
        'Rising Pune Supergiants': 'Rising Pune Supergiant'
    })

    # Filter season matches
    df = df[df['Season'] == season]
    if df.empty:
        return f"No matches for {season}"

    # Get teams
    teams = pd.concat([df['Team1'], df['Team2']]).unique()

    # Calculate stats
    stats = []
    for team in teams:
        matches = df[df['Team1'] == team].shape[0] + df[df['Team2'] == team].shape[0]
        wins = df[df['WinningTeam'] == team].shape[0]
        stats.append({'Team': team, 'Matches': matches, 'Wins': wins, 'Points': wins * 2})

    # Create and sort DataFrame
    result = pd.DataFrame(stats).set_index('Team').sort_values(['Points', 'Team'], ascending=[False, True])

    # Add Position as object type to allow mixed types
    result['Position'] = result['Points'].rank(method='first', ascending=False).astype('object')

    # Update positions for top 4 teams
    final = df[df['MatchNumber'] == 'Final']
    if not final.empty:
        winner = final['WinningTeam'].iloc[0]
        runner = final['Team1'].iloc[0] if final['Team2'].iloc[0] == winner else final['Team2'].iloc[0]
        result.at[winner, 'Position'] = 'Winner'
        result.at[runner, 'Position'] = 'Runner'

    q2 = df[df['MatchNumber'] == 'Qualifier 2']
    if not q2.empty:
        loser = q2['Team1'].iloc[0] if q2['Team2'].iloc[0] == q2['WinningTeam'].iloc[0] else q2['Team2'].iloc[0]
        result.at[loser, 'Position'] = 3

    elim = df[df['MatchNumber'] == 'Eliminator']
    if not elim.empty:
        loser = elim['Team1'].iloc[0] if elim['Team2'].iloc[0] == elim['WinningTeam'].iloc[0] else elim['Team2'].iloc[0]
        result.at[loser, 'Position'] = 4

    return result

# Print points table for 2016
print(point_table('2016'))
````
