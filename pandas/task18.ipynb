````python
import pandas as pd

# Load the dataset from the URL
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vT3D_x_4DS6d51LKJ7ze1sxT5WpV5uiSVOFYHLwBiGru6vFyVv5h5-83AwFjxWYiWfCDjDAaarHAV-k/pub?gid=0&single=true&output=csv"
df = pd.read_csv(url)

# Get unique teams
teams = df['Team'].unique()

# Initialize dictionary to store on-target percentages
on_target_percentages = {}

# Loop through each team
for team in teams:
    # Filter matches for the team
    team_matches = df[df['Team'] == team]
    # Sum total attempts and on-target shots
    total_attempts = team_matches['Total Attempts'].sum()
    on_target = team_matches['On Target'].sum()
    # Calculate percentage (avoid division by zero)
    if total_attempts > 0:
        percentage = (on_target / total_attempts) * 100
        # Round to 2 decimal places
        on_target_percentages[team] = round(percentage, 2)

# Sort dictionary by percentage values (descending order)
sorted_percentages = dict(sorted(on_target_percentages.items(), key=lambda x: x[1], reverse=True))

import pandas as pd
dff = pd.DataFrame.from_dict(sorted_percentages, orient='index', columns=['OnTarget%'])
dff
````

````python
import pandas as pd

# Load the IPL matches dataset
ipl_matches = "https://docs.google.com/spreadsheets/d/e/2PACX-1vRy2DUdUbaKx_Co9F0FSnIlyS-8kp4aKv_I0-qzNeghiZHAI_hw94gKG22XTxNJHMFnFVKsO4xWOdIs/pub?gid=1655759976&single=true&output=csv"
ipl_data = pd.read_csv(ipl_matches)

# Define team name mapping for standardization
team_name_mapping = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Rising Pune Supergiants': 'Rising Pune Supergiant'
}

# Replace old team names with current names
ipl_data['Team1'] = ipl_data['Team1'].replace(team_name_mapping)
ipl_data['Team2'] = ipl_data['Team2'].replace(team_name_mapping)
ipl_data['WinningTeam'] = ipl_data['WinningTeam'].replace(team_name_mapping)
ipl_data['TossWinner'] = ipl_data['TossWinner'].replace(team_name_mapping)

# Exclude no-result matches
ipl_data = ipl_data[ipl_data['WinningTeam'].notna()]

# Get unique teams
teams = pd.concat([ipl_data['Team1'], ipl_data['Team2']]).unique()

# Initialize lists to store statistics
team_stats = []

# Calculate statistics for each team
for team in teams:
    # Matches played (home or away)
    matches_played = len(ipl_data[(ipl_data['Team1'] == team) | (ipl_data['Team2'] == team)])
    
    # Total wins
    wins = len(ipl_data[ipl_data['WinningTeam'] == team])
    win_percentage = (wins / matches_played * 100) if matches_played > 0 else 0
    
    # Home matches (Team1) and home wins
    home_matches = len(ipl_data[ipl_data['Team1'] == team])
    home_wins = len(ipl_data[(ipl_data['Team1'] == team) & (ipl_data['WinningTeam'] == team)])
    home_win_percentage = (home_wins / home_matches * 100) if home_matches > 0 else 0
    
    # Away matches (Team2) and away wins
    away_matches = len(ipl_data[ipl_data['Team2'] == team])
    away_wins = len(ipl_data[(ipl_data['Team2'] == team) & (ipl_data['WinningTeam'] == team)])
    away_win_percentage = (away_wins / away_matches * 100) if away_matches > 0 else 0
    
    # Append statistics
    team_stats.append({
        'Team Name': team,
        'Matches Played': matches_played,
        'Win%': round(win_percentage, 2),
        'Home Win%': round(home_win_percentage, 2),
        'Away Win%': round(away_win_percentage, 2)
    })

# Create DataFrame
team_df = pd.DataFrame(team_stats)

# Sort by Win% in descending order
team_df = team_df.sort_values(by='Win%', ascending=False).reset_index(drop=True)

# Print the DataFrame
print("IPL Team Statistics (2008-2022, Sorted by Win%):")
print(team_df)
````

### `Q-8:` Player with most appearance in final match.

`Team1Players` and `Team2Players` have all players name. It is not a list of players name instead it is str. So handle it as string.

Hint: split and strip will help; Make a series of all players in final and do value counts


````python
import pandas as pd

# Load the IPL matches dataset
ipl_matches = "https://docs.google.com/spreadsheets/d/e/2PACX-1vRy2DUdUbaKx_Co9F0FSnIlyS-8kp4aKv_I0-qzNeghiZHAI_hw94gKG22XTxNJHMFnFVKsO4xWOdIs/pub?gid=1655759976&single=true&output=csv"
ipl_data = pd.read_csv(ipl_matches)

# Define team name mapping for standardization
team_name_mapping = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Rising Pune Supergiants': 'Rising Pune Supergiant'
}

# Replace old team names with current names
ipl_data['Team1'] = ipl_data['Team1'].replace(team_name_mapping)
ipl_data['Team2'] = ipl_data['Team2'].replace(team_name_mapping)
ipl_data['WinningTeam'] = ipl_data['WinningTeam'].replace(team_name_mapping)

# Filter final matches (where MatchNumber is 'Final')
final_matches = ipl_data[ipl_data['MatchNumber'] == 'Final']

# Initialize an empty list to store all player names from finals
all_players = []

# Process Team1Players and Team2Players
for _, row in final_matches.iterrows():
    # Handle Team1Players (string like "['Player1', 'Player2', ...]")
    team1_players = row['Team1Players']
    # Remove square brackets and split by comma
    team1_players = team1_players.strip("[]").split(", ")
    # Clean each player name (remove quotes and extra spaces)
    team1_players = [player.strip("' ") for player in team1_players]
    
    # Handle Team2Players similarly
    team2_players = row['Team2Players']
    team2_players = team2_players.strip("[]").split(", ")
    
    team2_players = [player.strip("' ") for player in team2_players]
    
    # Add all players to the list
    all_players.extend(team1_players + team2_players)

# Create a Series and count player appearances
player_counts = pd.Series(all_players).value_counts()
# Get the player(s) with the most appearances
max_appearances = player_counts.max()
top_players = player_counts[player_counts == max_appearances]

# Print the result
print("Player(s) with the most appearances in IPL finals (2008-2022):")
print(top_players)
````

### `Q-9:` IPL Point Table

Make a function `point_table` which take `season` as parameter and show points table in non-ascendng order of points and in ascending order of team name.

For winning - 2 Ponits;
For loosing - 0 Point
For not result both team gets 1 points.

Make dataframe which will have
`TeamName` `MatchesPlayed` `MatchesWon` `NoResult` `Points`
make `TeamName` as index.

```
season parametr should be one of these->
['2022', '2021', '2020/21', '2019', '2018', '2017', '2016', '2015', '2014', '2013', '2012', '2011', '2009/10', '2009', '2007/08']
```


````python
import pandas as pd

def point_table(season):
    """
    Generate IPL points table for a given season.
    
    Parameters:
    season (str): Season to analyze (e.g., '2022', '2021', '2020/21', etc.)
    
    Returns:
    DataFrame: Points table with TeamName as index, sorted by Points (desc) and TeamName (asc)
    """
    # Load the IPL matches dataset
    ipl_matches = "https://docs.google.com/spreadsheets/d/e/2PACX-1vRy2DUdUbaKx_Co9F0FSnIlyS-8kp4aKv_I0-qzNeghiZHAI_hw94gKG22XTxNJHMFnFVKsO4xWOdIs/pub?gid=1655759976&single=true&output=csv"
    ipl_data = pd.read_csv(ipl_matches)

    # Define team name mapping for standardization
    team_name_mapping = {
        'Delhi Daredevils': 'Delhi Capitals',
        'Kings XI Punjab': 'Punjab Kings',
        'Rising Pune Supergiants': 'Rising Pune Supergiant'
    }

    # Replace old team names with current names
    ipl_data['Team1'] = ipl_data['Team1'].replace(team_name_mapping)
    ipl_data['Team2'] = ipl_data['Team2'].replace(team_name_mapping)
    ipl_data['WinningTeam'] = ipl_data['WinningTeam'].replace(team_name_mapping)

    # Validate season
    valid_seasons = ['2022', '2021', '2020/21', '2019', '2018', '2017', '2016', '2015', 
                     '2014', '2013', '2012', '2011', '2009/10', '2009', '2007/08']
    if season not in valid_seasons:
        print(f"Invalid season. Choose from: {valid_seasons}")
        return None

    # Filter matches for the specified season
    season_data = ipl_data[ipl_data['Season'] == season]

    if season_data.empty:
        print(f"No matches found for season {season}.")
        return None

    # Get unique teams for the season
    teams = pd.concat([season_data['Team1'], season_data['Team2']]).unique()

    # Initialize lists to store statistics
    team_stats = []

    # Calculate statistics for each team
    for team in teams:
        # Matches played (home or away)
        matches_played = len(season_data[(season_data['Team1'] == team) | (season_data['Team2'] == team)])
        
        # Matches won
        matches_won = len(season_data[season_data['WinningTeam'] == team])
        
        # No result matches (WinningTeam is NaN)
        no_result_matches = len(season_data[
            ((season_data['Team1'] == team) | (season_data['Team2'] == team)) & 
            (season_data['WinningTeam'].isna())
        ])
        
        # Calculate points: 2 for win, 0 for loss, 1 for no result
        points = (matches_won * 2) + (no_result_matches * 1)
        
        # Append statistics
        team_stats.append({
            'TeamName': team,
            'MatchesPlayed': matches_played,
            'MatchesWon': matches_won,
            'NoResult': no_result_matches,
            'Points': points
        })

    # Create DataFrame
    points_table_df = pd.DataFrame(team_stats)

    # Set TeamName as index
    points_table_df.set_index('TeamName', inplace=True)

    # Sort by Points (descending) and TeamName (ascending)
    points_table_df = points_table_df.sort_values(by=['Points', 'TeamName'], 
                                                 ascending=[False, True])

    # Return the DataFrame
    return points_table_df

# Example usage
season = '2016'
print(f"IPL Points Table for Season {season}:")
result = point_table(season)
if result is not None:
    print(result)
````

Extend the above IPL Point Table with an extra column as `SeasonPosition`

Team below top 4 after sorting on `Points` and then on `TeamName` Will have same `SeasonPosition` as there rank. use rank function.

Teams in Top four will have `SeasonPosition` as:
```
    'Winner' - Team won final
    'Runner' - Team lost Final
    3 - Losing Team in Qualifier2
    4 - Losing Team in Eliminator
```

For changing value of pariticular cell use `df.at[row_index, col_label] = value`

Output of two Top 2 in season 2022. Your result should have all teams.
```
TeamName    MatchesPlayed	MatchesWon	NoResult	Points   SeasonPosition

Gujarat Titans	    16	   12	       0	     24         Winner
Rajasthan Royals	  17	   10	       0	     20         Runner

```

Note: If you try to chnage value of view of any dataframe a warnig will be shown. To avoid it, make a copy of the dataframe you want to change in by `df.copy()`

````python
import pandas as pd

def point_table(season):
    """
    Generate IPL points table for a given season with SeasonPosition.
    
    Parameters:
    season (str): Season to analyze (e.g., '2022')
    
    Returns:
    DataFrame: Points table with TeamName as index, sorted by Points (desc) and TeamName (asc)
    """
    # Load the IPL matches dataset
    ipl_matches = "https://docs.google.com/spreadsheets/d/e/2PACX-1vRy2DUdUbaKx_Co9F0FSnIlyS-8kp4aKv_I0-qzNeghiZHAI_hw94gKG22XTxNJHMFnFVKsO4xWOdIs/pub?gid=1655759976&single=true&output=csv"
    ipl_data = pd.read_csv(ipl_matches)

    # Define team name mapping for standardization
    team_name_mapping = {
        'Delhi Daredevils': 'Delhi Capitals',
        'Kings XI Punjab': 'Punjab Kings',
        'Rising Pune Supergiants': 'Rising Pune Supergiant'
    }

    # Replace old team names with current names
    ipl_data['Team1'] = ipl_data['Team1'].replace(team_name_mapping)
    ipl_data['Team2'] = ipl_data['Team2'].replace(team_name_mapping)
    ipl_data['WinningTeam'] = ipl_data['WinningTeam'].replace(team_name_mapping)

    # Validate season
    valid_seasons = ['2022', '2021', '2020/21', '2019', '2018', '2017', '2016', '2015', 
                     '2014', '2013', '2012', '2011', '2009/10', '2009', '2007/08']
    if season not in valid_seasons:
        print(f"Invalid season. Choose from: {valid_seasons}")
        return None

    # Filter matches for the specified season
    season_data = ipl_data[ipl_data['Season'] == season]

    if season_data.empty:
        print(f"No matches found for season {season}.")
        return None

    # Get unique teams for the season
    teams = pd.concat([season_data['Team1'], season_data['Team2']]).unique()

    # Initialize lists to store statistics
    team_stats = []

    # Calculate statistics for each team
    for team in teams:
        # Matches played (home or away)
        matches_played = len(season_data[(season_data['Team1'] == team) | (season_data['Team2'] == team)])
        
        # Matches won
        matches_won = len(season_data[season_data['WinningTeam'] == team])
        
        # No result matches
        no_result_matches = len(season_data[
            ((season_data['Team1'] == team) | (season_data['Team2'] == team)) & 
            (season_data['WinningTeam'].isna())
        ])
        
        # Calculate points: 2 for win, 0 for loss, 1 for no result
        points = (matches_won * 2) + (no_result_matches * 1)
        
        # Append statistics
        team_stats.append({
            'TeamName': team,
            'MatchesPlayed': matches_played,
            'MatchesWon': matches_won,
            'NoResult': no_result_matches,
            'Points': points
        })

    # Create DataFrame
    points_table_df = pd.DataFrame(team_stats)

    # Set TeamName as index
    points_table_df.set_index('TeamName', inplace=True)

    # Sort by Points (descending) and TeamName (ascending)
    points_table_df = points_table_df.sort_values(by=['Points', 'TeamName'], 
                                                 ascending=[False, True])

    # Create a copy to avoid SettingWithCopyWarning
    points_table_df = points_table_df.copy()

    # Add SeasonPosition column (default to rank)
    points_table_df['SeasonPosition'] = points_table_df['Points'].rank(method='first', ascending=False).astype(int)

    # Identify top 4 teams' positions
    # Final match (Winner and Runner)
    final_match = season_data[season_data['MatchNumber'] == 'Final']
    if not final_match.empty:
        winner = final_match['WinningTeam'].iloc[0]
        runner = final_match['Team1'].iloc[0] if final_match['Team2'].iloc[0] == winner else final_match['Team2'].iloc[0]
        points_table_df.at[winner, 'SeasonPosition'] = 'Winner'
        points_table_df.at[runner, 'SeasonPosition'] = 'Runner'

    # Qualifier 2 (3rd place: loser)
    qualifier2 = season_data[season_data['MatchNumber'] == 'Qualifier 2']
    if not qualifier2.empty:
        q2_winner = qualifier2['WinningTeam'].iloc[0]
        q2_loser = qualifier2['Team1'].iloc[0] if qualifier2['Team2'].iloc[0] == q2_winner else qualifier2['Team2'].iloc[0]
        points_table_df.at[q2_loser, 'SeasonPosition'] = 3

    # Eliminator (4th place: loser)
    eliminator = season_data[season_data['MatchNumber'] == 'Eliminator']
    if not eliminator.empty:
        elim_winner = eliminator['WinningTeam'].iloc[0]
        elim_loser = eliminator['Team1'].iloc[0] if eliminator['Team2'].iloc[0] == elim_winner else eliminator['Team2'].iloc[0]
        points_table_df.at[elim_loser, 'SeasonPosition'] = 4

    return points_table_df

# Example usage for 2022
season = '2016'
print(f"IPL Points Table for Season {season}:")
result = point_table(season)
if result is not None:
    print(result)
````
