```python
from bs4 import BeautifulSoup
import time
import requests
import pandas as pd
import numpy as np
import re
import random

# Lists to hold the data
image_url = []
name = []
star_rating = []
rating = []
reviews = []
ram = []
rom = []
memory_card = []
display = []
battery = []
processor = []
warranty = []
price = []
discounted_price = []
actual_price = []
deliverydate = []
discountupto = []

# Loop through pages 1 to 41
for i in range(1, 42):
    print(f"Scraping Page {i}...")
    try:
        webpage = requests.get(
            f'https://www.flipkart.com/search?q=smartphones&page={i}',
            headers={
                'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36'
            }
        ).text
        soup = BeautifulSoup(webpage, 'lxml')
        all_divs = soup.find_all('div', class_="tUxRFH")

        for phone in all_divs:
            try:
                image_url.append(phone.find("img", class_="DByuf4")['src'])
            except:
                image_url.append(np.nan)

            try:
                name.append(phone.find("div", class_="KzDlHZ").text.strip())
            except:
                name.append(np.nan)

            try:
                star_rating.append(phone.find("div", class_="XQDdHH").text.strip())
            except:
                star_rating.append(np.nan)

            try:
                rating_review_text = phone.find_all("span", class_='Wphh3N')[0].find("span").text.strip()
                numbers = re.findall(r'[\d,]+', rating_review_text)
                rating.append(numbers[0] if len(numbers) > 0 else np.nan)
                reviews.append(numbers[1] if len(numbers) > 1 else np.nan)
            except:
                rating.append(np.nan)
                reviews.append(np.nan)

            try:
                specs = phone.find("ul", class_="G4BRas").find_all("li", class_="J+igdf")
                rom_ram_text = specs[0].text.strip() if len(specs) > 0 else ""
                parts = [x.strip() for x in rom_ram_text.split('|')]
                ram.append(parts[0] if len(parts) > 0 else np.nan)
                rom.append(parts[1] if len(parts) > 1 else np.nan)
                memory_card.append(parts[2] if len(parts) > 2 else np.nan)
            except:
                ram.append(np.nan)
                rom.append(np.nan)
                memory_card.append(np.nan)

            try:
                display.append(specs[1].text.strip() if len(specs) > 1 else np.nan)
            except:
                display.append(np.nan)

            try:
                battery.append(specs[3].text.strip() if len(specs) > 3 else np.nan)
            except:
                battery.append(np.nan)

            try:
                processor.append(specs[4].text.strip() if len(specs) > 4 else np.nan)
            except:
                processor.append(np.nan)

            try:
                warranty.append(specs[5].text.strip() if len(specs) > 5 else np.nan)
            except:
                warranty.append(np.nan)

            try:
                discounted_price.append(phone.find("div", class_="Nx9bqj _4b5DiR").text.strip())
            except:
                discounted_price.append(np.nan)

            try:
                actual_price.append(phone.find("div", class_="yRaY8j ZYYwLA").text.strip())
            except:
                actual_price.append(np.nan)

            try:
                deliverydate.append(phone.find_all("div", class_="yiggsN")[1].text.strip())
            except:
                deliverydate.append(np.nan)

            try:
                discountupto.append(phone.find("div", class_="UkUFwK").find('span').text.strip())
            except:
                discountupto.append(np.nan)

        # Random sleep between 1 and 3 seconds
        delay = round(random.uniform(1, 3), 2)
        print(f"Sleeping for {delay} seconds...")
        time.sleep(delay)

    except Exception as e:
        print(f"Error on page {i}: {e}")
        continue

# Create dictionary of all columns
columns = {
    'Name': name,
    'Image URL': image_url,
    'Star Rating': star_rating,
    'Ratings': rating,
    'Reviews': reviews,
    'RAM': ram,
    'ROM': rom,
    'Expandable Memory': memory_card,
    'Display': display,
    'Battery': battery,
    'Processor': processor,
    'Warranty': warranty,
    'Price': price,
    'Discounted Price': discounted_price,
    'Actual Price': actual_price,
    'Delivery Date': deliverydate,
    'Discount Upto': discountupto
}

# Normalize all columns to same length
max_len = max(len(v) for v in columns.values())
for k, v in columns.items():
    while len(columns[k]) < max_len:
        columns[k].append(np.nan)

# Convert to DataFrame
df = pd.DataFrame(columns)

# Clean currency fields
df['Actual Price'] = df['Actual Price'].str.replace('', '', regex=False).str.replace(',', '', regex=False)
df['Discounted Price'] = df['Discounted Price'].str.replace('', '', regex=False).str.replace(',', '', regex=False)
df['Discount Upto'] = df['Discount Upto'].str.replace('', '', regex=False).str.replace(',', '', regex=False)

# Save to CSV
df.to_csv("flipkart_smartphones.csv", index=False)
print("Scraping completed and data saved.")
```

```python
from bs4 import BeautifulSoup
import time
import requests
import pandas as pd
import numpy as np
import re
import random

# Data lists
image_url = []
name = []
star_rating = []
rating = []
reviews = []
ram = []
rom = []
memory_card = []
display = []
battery = []
processor = []
warranty = []
discounted_price = []
actual_price = []
deliverydate = []
discountupto = []

# Scrape 41 pages
for i in range(1, 42):
    print(f"\n Scraping Page {i}...")
    try:
        response = requests.get(
            f'https://www.flipkart.com/search?q=smartphones&page={i}',
            headers={'User-Agent': 'Mozilla/5.0'}
        )
        soup = BeautifulSoup(response.text, 'lxml')
        all_divs = soup.find_all('div', class_="tUxRFH")

        for phone in all_divs:
            # Image
            img = phone.find("img", class_="DByuf4")
            image_url.append(img['src'] if img else np.nan)

            # Name
            title = phone.find("div", class_="KzDlHZ")
            name.append(title.text.strip() if title else np.nan)

            # Star Rating
            star = phone.find("div", class_="XQDdHH")
            star_rating.append(star.text.strip() if star else np.nan)

            # Ratings and Reviews
            try:
                rating_text = phone.find_all("span", class_='Wphh3N')[0].find("span").text.strip()
                nums = re.findall(r'[\d,]+', rating_text)
                rating.append(nums[0].replace(',', '') if len(nums) > 0 else np.nan)
                reviews.append(nums[1].replace(',', '') if len(nums) > 1 else np.nan)
            except:
                rating.append(np.nan)
                reviews.append(np.nan)

            # Specs
            try:
                specs = phone.find("ul", class_="G4BRas").find_all("li", class_="J+igdf")
                parts = specs[0].text.strip().split('|') if specs else []
                ram.append(parts[0].strip() if len(parts) > 0 else np.nan)
                rom.append(parts[1].strip() if len(parts) > 1 else np.nan)
                memory_card.append(parts[2].strip() if len(parts) > 2 else np.nan)
                display.append(specs[1].text.strip() if len(specs) > 1 else np.nan)
                battery.append(specs[3].text.strip() if len(specs) > 3 else np.nan)
                processor.append(specs[4].text.strip() if len(specs) > 4 else np.nan)
                warranty.append(specs[5].text.strip() if len(specs) > 5 else np.nan)
            except:
                ram.append(np.nan)
                rom.append(np.nan)
                memory_card.append(np.nan)
                display.append(np.nan)
                battery.append(np.nan)
                processor.append(np.nan)
                warranty.append(np.nan)

            # Prices
            dp = phone.find("div", class_="Nx9bqj _4b5DiR")
            discounted_price.append(dp.text.strip() if dp else np.nan)

            ap = phone.find("div", class_="yRaY8j ZYYwLA")
            actual_price.append(ap.text.strip() if ap else np.nan)

            # Delivery
            try:
                deliverydate.append(phone.find_all("div", class_="yiggsN")[1].text.strip())
            except:
                deliverydate.append(np.nan)

            # Discount
            try:
                discount = phone.find("div", class_="UkUFwK").find("span")
                discountupto.append(discount.text.strip() if discount else np.nan)
            except:
                discountupto.append(np.nan)

        # Delay to avoid blocking
        sleep_time = round(random.uniform(1.5, 3), 2)
        print(f"Sleeping for {sleep_time}s...")
        time.sleep(sleep_time)

    except Exception as e:
        print(f" Error on page {i}: {e}")
        continue

# Create DataFrame
df = pd.DataFrame({
    'Name': name,
    'Image URL': image_url,
    'Star Rating': star_rating,
    'Ratings': rating,
    'Reviews': reviews,
    'RAM': ram,
    'ROM': rom,
    'Expandable Memory': memory_card,
    'Display': display,
    'Battery': battery,
    'Processor': processor,
    'Warranty': warranty,
    'Discounted Price': discounted_price,
    'Actual Price': actual_price,
    'Delivery Date': deliverydate,
    'Discount Upto': discountupto
})

# Clean currency fields
df['Actual Price'] = df['Actual Price'].str.replace('', '', regex=False).str.replace(',', '', regex=False)
df['Discounted Price'] = df['Discounted Price'].str.replace('', '', regex=False).str.replace(',', '', regex=False)
df['Discount Upto'] = df['Discount Upto'].str.replace('% off', '', regex=False).str.strip()

# Convert prices to numeric
df['Actual Price'] = pd.to_numeric(df['Actual Price'], errors='coerce')
df['Discounted Price'] = pd.to_numeric(df['Discounted Price'], errors='coerce')
df['Discount Upto'] = pd.to_numeric(df['Discount Upto'], errors='coerce')

# Save to CSV
df.to_csv("flipkart_smartphones_cleaned.csv", index=False)
print("\n Scraping completed and data saved to 'flipkart_smartphones_cleaned.csv'")
```
