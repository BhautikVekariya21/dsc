```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
cwc_deliveries = pd.read_csv('cwc2023.csv')
cwc_matches = pd.read_csv('cwcmatches2023.csv')
cwc_deliveries
```

```python
#filling na values for all float datatypes
cwc_deliveries[["wides", "noballs", "byes", "legbyes", "penalty"]] = cwc_deliveries[["wides", "noballs", "byes", "legbyes", "penalty"]].fillna(0)
```

```python
cwc_deliveries['start_date'] = pd.to_datetime(cwc_deliveries['start_date'])
```

```python
#Changing the datatype to Integer 
cwc_deliveries['penalty'] = cwc_deliveries['penalty'].astype(np.int16)
cwc_deliveries['wides'] = cwc_deliveries['wides'].astype(np.int32)
cwc_deliveries['innings'] = cwc_deliveries['innings'].astype(np.int16)
cwc_deliveries['runs_off_bat'] = cwc_deliveries['runs_off_bat'].astype(np.int16)
cwc_deliveries['extras'] = cwc_deliveries['extras'].astype(np.int16)
cwc_deliveries['wides'] = cwc_deliveries['wides'].astype(np.int16)
cwc_deliveries['noballs'] = cwc_deliveries['noballs'].astype(np.int16)
cwc_deliveries['byes'] = cwc_deliveries['byes'].astype(np.int16)
cwc_deliveries['legbyes'] = cwc_deliveries['legbyes'].astype(np.int16)
```

```python
print(cwc_deliveries['penalty'].unique())
print(cwc_deliveries['wicket_type'].unique())
print(cwc_deliveries['other_wicket_type'].unique())
print(cwc_deliveries['other_player_dismissed'].unique())
```

```python
cwc_deliveries['wicket_type'] = cwc_deliveries['wicket_type'].fillna("No Wicket")
cwc_deliveries['player_dismissed'] = cwc_deliveries['player_dismissed'].fillna("No Wicket")
```

```python
print(cwc_deliveries.isnull().sum())
print(cwc_deliveries.isnull().sum())
```

```python
unique_counts = cwc_deliveries['player_dismissed'].value_counts()
```

```python
filtered_counts = unique_counts[unique_counts > 8]
filtered_counts
```

```python
#Creating wicket column for each delivery, 1 if its a wicket or 0 for non dismissal
cwc_deliveries['wicket'] = cwc_deliveries['player_dismissed'].apply(lambda x: 0 if x == "No Wicket" else 1)
```

```python
cwc_deliveries.isnull().sum()
cwc_matches.isna().sum()
```

```python
cwc_matches.rename(columns={"winner_runs":"win_by_runs", "winner_wickets":"win_by_wickets"}, inplace=True)
```

```python
cwc_matches['win_by_runs'] = cwc_matches['win_by_runs'].fillna(0)
cwc_matches['win_by_wickets'] = cwc_matches['win_by_wickets'].fillna(0)
```

```python
cwc_matches = cwc_matches.drop(["reserve_umpire"],axis =1)
```

```python
#Lets create overs, ball number and balls left columns
cwc_deliveries["ball"] = cwc_deliveries["ball"].astype(str)
cwc_deliveries[["over","ball_number"]] = cwc_deliveries["ball"].str.split(".",expand = True).astype(int)
```

```python
cwc_deliveries
```

```python
cwc_deliveries["over"] = cwc_deliveries["over"]+1
```

```python
#Removing columns
cwc_deliveries = cwc_deliveries.drop(['other_player_dismissed', 'other_wicket_type'], axis =1)
```

```python

#Create total run per ball
#creating total runs column to calculate runs on each ball incusive of extras
cwc_deliveries["total_runs_per_ball"] = cwc_deliveries["runs_off_bat"] + cwc_deliveries["extras"]
```

```python
#Create cumulative_runs columns to calculate cumulative runs based on Innings 
cwc_deliveries["cum_runs"] = cwc_deliveries.groupby(["match_id", "innings"])["total_runs_per_ball"].cumsum()
```

```python
#Split the dataframe into first innings and second innings to calculate target and required run rate(rrr)
first_innings_df = cwc_deliveries[cwc_deliveries["innings"] == 1]
first_innings = first_innings_df.groupby("match_id")["cum_runs"].max().reset_index()
first_innings.rename(columns={"cum_runs": "target"}, inplace=True)
```

```python
first_innings
```

```python
second_innings_df = cwc_deliveries[cwc_deliveries["innings"] == 2]
second_innings_df = second_innings_df.merge(first_innings, on="match_id", how="left")
second_innings_df["target"] = second_innings_df["target"]+1
```

```python
#Calculate runs needed and rrr for second innings dataframe
second_innings_df["runs_needed"] = second_innings_df["target"] - second_innings_df["cum_runs"]
second_innings_df.head(10)
```

```python
second_innings_df.columns
```

```python
#Combine both datasets to creata a final cleaned dataframe
full_innings_df = pd.concat([first_innings_df,second_innings_df])
```

```python
full_innings_df.fillna(0,inplace = True)
```

```python
cwc_matches.rename(columns = {"match_number":"match_id", "start_date":"date"},inplace = True)
```

```python
#Merge matches data with innings to get the winner, win_by_runs and win_by_wickets data
full_innings_df = full_innings_df.merge(cwc_matches[["match_id","winner", "win_by_runs", "win_by_wickets","match_type","toss_winner","toss_decision"]], on = "match_id")
```

```python
#Define result of the match
full_innings_df["result"] = full_innings_df.apply(lambda x: 1 if x["batting_team"] == x["winner"] else 0, axis=1)
```
