```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
sns.set_theme(style="dark")
plt.style.use("dark_background")
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# scatter plot -> axes plot function
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time')
```

```python
sns.set_style("dark")
sns.set_theme(style="dark")
plt.style.use("dark_background")
# Sample plot
data = sns.load_dataset("iris")
sns.scatterplot(x="sepal_length", y="sepal_width", data=data)

plt.show()
```

```python
#replot -> figure level -> square shape
sns.relplot(
    data=tips, 
    x="total_bill", 
    y="total_bill", 
    kind="scatter", 
    hue="sex", 
    style="time", 
    size="size"
)
```

```python
sns.relplot(
    data=tips, 
    x="total_bill", 
    y="tip", 
    kind="scatter", 
    hue="sex", 
    style="time", 
    size="size"
)
```

```python
# line plot 
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
#axes level function
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

```python
# using relpplot
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
# hue -> style
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df
```

```python
sns.relplot(kind='line',data=temp_df,x='year',y='lifeExp',hue='country',style='continent',size='continent')
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp',hue='country')
```

```python
# facet plot -> figure level funnction -> work with replot
# it will not work with with scatterplot and lineplot
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',row='sex',kind='scatter')
```
